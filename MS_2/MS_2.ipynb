## Задание 4.1

Пусть $x_1, ..., x_n$ ∼ $U([-3\theta_0, \theta_0])$. Пусть $\hat{\theta}_n = 4x_{[n]} + x_{[1]}$ - оценка для параметра $\theta_0$.

- Исследуйте $\hat{\theta}_n$ на (асимптотическую) несмещенность;
- Исследуйте $\hat{\theta}_n$ на состоятельность.

### Решение:

```python
import matplotlib.pyplot as plt
from scipy import stats as st
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

1. Для того, чтобы исследовать на несмещенность, предлагается:

- выбрать несколько значений параметра $\theta_0$, например, $\theta_0 \in \{1, 2, 3, 4\}$;
- для каждого $\theta_0$ на отрезке $[-3\theta_0, \theta_0]$ сгенерировать $n$ выборок из равномерного распределения из $k$ элементов, где $n = 100, k \in \{100, 1000, 10000\}$;
- для каждой такой выборки вычислить оценку параметра $\hat{\theta}_n = 4x_{[n]} + x_{[1]}$, после усреднить все вычисленные оценки и сравнить с самим параметром $\theta_0$.

```python
theta = [1, 2, 3, 4]
n = 100
k = [10 ** i for i in range(2, 5)]
s = len(str(k[-1])) * 2
for theta_0 in theta:
    print(f'Параметр theta_0 = {theta_0}')
    # распределние задается на отрезке [loc, loc + scale]
    loc = -3 * theta_0
    scale = 4 * theta_0
    print(f'Выборка генерируется на отрезке [{loc}, {loc + scale}]')
    print(f'Отклонение среднего значения по n выборкам оценки параметра \nот самого параметра:')
    th = []
    d = []
    for sample_size in k:
        for _ in range(n):
            ans = []
            sample = st.uniform.rvs(loc=loc, scale=scale, size=sample_size)
            theta_n = 4 * max(sample) + min(sample)
            ans.append(theta_n)
        th.append(round(np.mean(ans), 3))
        d.append(round((th[-1] - theta_0) / theta_0 * 100, 2))
    print('-' * (3 * s + 7))
    print(' ' * (s - 1) + 'k | ' + ' ' * (s - 9) + 'E theta_n | ' + ' ' * (s - 3) + 'diff')
    print('-' * (3 * s + 7))
    for i in range(len(k)):
        print(f'{k[i]:>{s}} | {th[i]:>{s}} | {d[i]:>{s}}%')
    print('-' * (3 * s + 7))
    print()
```

Сразу же видно, что математическое ожидание оценки параметра $\mathbb{E} \hat{\theta}_n \neq \theta_0$. Однако при стремлении количество элементов выборки к бесконечности математическое ожидание стремится к значению параметра, т.е. $\mathbb{E} \hat{\theta}_n \xrightarrow{p} \theta_0$. Значит, данная оценка является **асимптотически несмещенной**.

2. Для того, чтобы исследовать на состоятельность, предлагается:

- выбрать несколько значений параметра $\theta_0$, например, $\theta_0 \in \{1, 2, 3, 4\}$;
- для каждого $\theta_0$ на отрезке $[-3\theta_0, \theta_0]$ сгенерировать выборку из равномерного распределения из $n$ элементов, где $n \in \{100, 1000, 10000, 100000, 1000000\}$;
- для каждой такой выборки выяснить, как сильно оценка параметра $\hat{\theta}_n = 4x_{[n]} + x_{[1]}$ отклоняется от самого параметра $\theta_0$.

```python
theta = [1, 2, 3, 4]
n = [10 ** k for k in range(2, 7)]
s = len(str(n[-1]))
for theta_0 in theta:
    ans = []
    th = []
    print(f'Параметр theta_0 = {theta_0}')
    # распределние задается на отрезке [loc, loc + scale]
    loc = -3 * theta_0
    scale = 4 * theta_0
    print(f'Выборка генерируется на отрезке [{loc}, {loc + scale}]')
    print('Зависимость отклонения оценки параметра \nот самого параметра для разных размеров выборки:')
    for sample_size in n:
        sample = st.uniform.rvs(loc=loc, scale=scale, size=sample_size)
        theta_n = 4 * max(sample) + min(sample)
        delta = round((theta_n - theta_0) / theta_0 * 100, 2)
        th.append(round(theta_n, 3))
        ans.append(delta)
    print('-' * (3 * s + 7))
    print(' ' * (s - 1) + 'n | ' + ' ' * (s - 7) + 'theta_n | ' + ' ' * (s - 3) + 'diff')
    print('-' * (3 * s + 7))
    for i in range(len(n)):
        print(f'{n[i]:>{s}} | {th[i]:>{s}} | {ans[i]:>{s}}%')
    print('-' * (3 * s + 7))
    print()
```

Для каждого рассмотренного отрезка с увеличением числа элементов в выборке отклонение оценки параметра $\hat{\theta}_n$ от самого параметра $\theta_0$ стремится к нулю, а значит, оценка параметра стремится по вероятности к самому параметру, т.е. $\hat{\theta}_n \xrightarrow{p} \theta_0$. Иными словами, такая **оценка является состоятельной**.
