### Задание 3.3

Приведите пример распределения, для которого центральная предельная теорема не выполняется. Продемонстрируйте это.

### Решение

Подберем какое-нибудь распределение, чьи характеристики не удовлетворяют условиям ЦПТ.

Возьмем, например, распределение Коши с несуществующим математическим ожиданием и бесконечной дисперсией.

```python
import matplotlib.pyplot as plt
from scipy import stats as st
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

Генерируем несколько выборок размера *sample_size* из распределения Коши. Для каждой такой выборки считаем среднее и смотрим, как они распределены. Если бы ЦПТ выполнялась, то мы увидели бы нормальное распределение. Что же мы увидим здесь?..

Построим несколько графиков, для каждого из которых выберем разное количество генерируемых выборок из n_samples.

```python
sample_size = 10000
n_samples = [50, 100, 200, 400, 800, 1500, 2000, 3000, 5000]
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, figsize=(15, 15), sharex=False)
axs = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9]

for i in range(9):
    samp = [np.mean(st.cauchy.rvs(size=sample_size)) for _ in range(n_samples[i])]
    sns.distplot(samp, kde=True, ax=axs[i])
    axs[i].set(title=f'Количество выборочных средних n = {n_samples[i]}',
              xlabel='x', ylabel=r'$p_{\xi}(x)$')
```

Несмотря на то, что при малых значениях количества выборочных средних распределения Коши, графики плотности распределения выборочных средних имеют куполообразную симметричную форму, при устремлении n к бесконечности распределение перестает хоть сколько-нибудь напоминать нормальное. Большое количество значений распределено вблизи нуля, однако распределение становится сильно несимметричным, причем большое количество значений попадает в длинные хвосты.

Данный смоделированный эксперимент показывает, что распределение Коши, не удовлетворяющее условиям ЦПТ в теории, действительно не удовлетворяет ей и на практике.
